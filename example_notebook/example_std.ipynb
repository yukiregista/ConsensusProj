```python
from importlib.resources import files
```

```python
EXAMPLE_TREES_PATH = files("Consensus.example_data").joinpath("std_ex.tre")
```

## Load Data

```python
from Consensus import *
```

```python
input_trees = TreeList_with_support.get(path = EXAMPLE_TREES_PATH, schema = "newick") # for files containing multiple trees
```

```python
stdg = STDGreedyConsensus(input_trees)
star_tree = Tree_with_support.get(data="(A,B,C,D,E,F,H,G);", schema="newick", taxon_namespace=stdg.taxon_namespace)
stdg.specify_initial_tree(star_tree) # starting from majority tree
stdg.greedy(method="first", order="BS")
```

```python
t = stdg.return_current_tree()
```
